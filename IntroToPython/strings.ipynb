# strings.ipynb
Very simple examples of common Python string operations.

## 1) Create sample variables
We define simple values that will be reused in later examples.

```python
# Sample variables
name = "Ana"
age = 14

single_word = "Python"
padded_text = "   hello world   "

numeric_str = "123"
binary_str = "1010"
hex_str = "1A"

words = ["red", "green", "blue"]

search_text = "banana"
slice_text = "Programming"
case_text = "python string methods are useful"

str1 = "Hello"
str2 = "World"
```

## 2) String literal
Single-quoted and double-quoted string literals.

```python
single_quoted = 'I am single-quoted'
double_quoted = "I am double-quoted"

print(single_quoted)
print(double_quoted)
```

## 3) Multi-String (Multiline String)
Use triple quotes for multiple lines.

```python
multi_line = """Line 1
Line 2
Line 3"""

print(multi_line)
```

## 4) String interpolation
Example with f-string and `str.format()`.

```python
# f-string
msg_f = f"My name is {name} and I am {age} years old."
print(msg_f)

# str.format()
msg_format = "My name is {} and I am {} years old.".format(name, age)
print(msg_format)
```

## 5) `bytes()`
Convert text to bytes and convert a list of integers to bytes.

```python
text = "ABC"
text_bytes = bytes(text, "utf-8")
print(text_bytes)

int_list_bytes = bytes([65, 66, 67])
print(int_list_bytes)
```

## 6) `index()` and `find()`
`find()` returns `-1` when not found.  
`index()` raises an error when not found.

```python
print("Text:", search_text)

# Found case
print("find('na'):", search_text.find("na"))
print("index('na'):", search_text.index("na"))

# Not found case
print("find('xy'):", search_text.find("xy"))

try:
    print("index('xy'):", search_text.index("xy"))
except ValueError as e:
    print("index('xy') raised ValueError:", e)
```

## 7) Slicing with `str[start:end]`
Examples with positive and negative indexes.

```python
print("Original:", slice_text)

print("First 3 chars:", slice_text[:3])
print("Middle (3:8):", slice_text[3:8])
print("Last 3 chars:", slice_text[-3:])
```

## 8) `strip()`, `lstrip()`, `rstrip()`
Remove spaces from both sides, left side, or right side.

```python
print(f"Original: [{padded_text}]")
print(f"strip():  [{padded_text.strip()}]")
print(f"lstrip(): [{padded_text.lstrip()}]")
print(f"rstrip(): [{padded_text.rstrip()}]")
```

## 9) `upper()`, `lower()`, `title()`, `capitalize()`
Case conversion methods.

```python
print("Original:   ", case_text)
print("upper():    ", case_text.upper())
print("lower():    ", case_text.lower())
print("title():    ", case_text.title())
print("capitalize():", case_text.capitalize())
```

## 10) `int(string)` and `int(string, base)`
Convert decimal, binary, and hexadecimal strings to integers.

```python
dec_value = int(numeric_str)
bin_value = int(binary_str, 2)
hex_value = int(hex_str, 16)

print(f"int('{numeric_str}') =", dec_value)
print(f"int('{binary_str}', 2) =", bin_value)
print(f"int('{hex_str}', 16) =", hex_value)
```

## 11) `str1 + str2`
Simple concatenation with `+`.

```python
concat_no_space = str1 + str2
concat_with_space = str1 + " " + str2

print("Without space:", concat_no_space)
print("With space:   ", concat_with_space)
```

## 12) `str.join([s1, s2, ...])`
Join a list of strings with different separators.

```python
print("Words:", words)
print("' '.join(words):", " ".join(words))
print("', '.join(words):", ", ".join(words))
print("'-'.join(words):", "-".join(words))
```
